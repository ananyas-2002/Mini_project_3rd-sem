```python
import re
```

```python
def preprocess_tokenize(data):
      
    text = str(data)
    text = re.sub(r'(\d+)', r'', data)
    
    text = text.replace('\n', '')
    text = text.replace('\r', '')
    text = text.replace('\t', '')
    text = text.replace('\u200d', '')
    text=re.sub(r"(__+)", ' ', str(data)) 
    text=re.sub(r"(--+)", ' ', str(data)) 
    text=re.sub(r"(~~+)", ' ', str(data))  
    text=re.sub(r"(\+\++)", ' ', str(data))   
    text=re.sub(r"(\.\.+)", ' ', str(data))  
        
    text=re.sub(r"[<>()|&©@#ø\[\]\'\",;:?.~*!]", ' ', str(data)) #remove <>()|&©ø"',;?~*!
    text = re.sub(r"([a-zA-Z])",' ',str(data))
    text = re.sub(r"(\s+)",' ',str(data))
    
    return data
```
